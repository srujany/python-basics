3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.



```python
# prompt: 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data for categories and values
data = {'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
        'Value': [10, 15, 12, 11, 16, 13, 9, 14, 14]}
df = pd.DataFrame(data)

# Create a box plot to compare categories based on values
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Comparison of Categories based on Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Create a bar plot to compare categories based on the mean values
plt.figure(figsize=(8,6))
sns.barplot(x='Category', y='Value', data=df, ci=None) # ci=None removes confidence interval
plt.title('Mean Value Comparison across Categories')
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
# prompt: 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

# Generate a synthetic dataset with correlated features
np.random.seed(0)
n_samples = 100
feature_1 = np.random.normal(0, 1, n_samples)
feature_2 = feature_1 + np.random.normal(0, 0.5, n_samples)  # Introduce correlation
feature_3 = -feature_1 + np.random.normal(0, 0.8, n_samples) # Introduce negative correlation
data = {'Feature 1': feature_1, 'Feature 2': feature_2, 'Feature 3': feature_3}
df = pd.DataFrame(data)


# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.



```python
# prompt: 1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Output the plot in the notebook
output_notebook()

# Create the figure
p = figure(width=600, height=400, title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

# Generate the x-values
x = np.linspace(0, 10, 100)

# Calculate the y-values (sine of x)
y = np.sin(x)

# Add the sine wave to the plot
p.line(x, y, line_width=2, color='navy')

# Show the plot
show(p)
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the

markers based on the 'sizes' and 'colors' columns.



```python
# prompt: 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

# Output the plot in the notebook
output_notebook()

# Create a ColumnDataSource
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
sizes = np.random.randint(5, 20, n)
colors = ["#%02x%02x%02x" % (r, g, 150) for r, g in zip(np.random.randint(0, 255, n), np.random.randint(0, 255, n))]

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create the figure
p = figure(width=600, height=400, title="Scatter Plot with varying sizes and colors", x_axis_label="x", y_axis_label="y")

# Add the scatter plot
p.circle(x='x', y='y', size='sizes', color='colors', source=source)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits [Apples', 'Oranges', 'Bananas', 'Pears']

counts [20, 25, 38, 35]



```python
# prompt: 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits [Apples', 'Oranges', 'Bananas', 'Pears']
# counts [20, 25, 38, 35]

# Output the plot in the notebook
output_notebook()

# Create a ColumnDataSource
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 38, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the figure
p = figure(x_range=fruits, width=600, height=400, title="Fruit Counts", x_axis_label="Fruit", y_axis_label="Count")

# Add the bar chart
p.vbar(x='fruits', top='counts', width=0.8, source=source, legend_field="fruits")

# Customize the plot (optional)
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
# prompt: 4. Create a Bokeh histogram to visualize the distribution of the given data.
# data hist np.random.randn(1688)
# hist, edges np.histogram(data hist, bins=30)

# Output the plot in the notebook
output_notebook()

# Generate the data
data_hist = np.random.randn(1688)
hist, edges = np.histogram(data_hist, bins=30)

# Create the figure
p = figure(width=600, height=400, title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Add the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")

# Show the plot
show(p)
```
